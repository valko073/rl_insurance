```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('logs/outfile-2019-07-05-14-44-59.p', 'rb') as fp:
    data= pickle.load(fp)
```

```python
data
```

```python
low_lim = 28995+1000
high_lim = low_lim + 25
```

```python
insurance_costs = np.array([x[1][0][0] for x in data[low_lim: high_lim]])
```

```python
agent_actions = [1 if x[1][1] in [2,3] else 0 for x in data[low_lim: high_lim]]
```

```python
plt.figure(figsize=(20., 4.8))
plt.plot(range(len(insurance_costs)),insurance_costs)
plt.title('Insurance cost')
plt.figure(figsize=(20., 4.8))
plt.plot(range(len(insurance_costs)), agent_actions)
plt.title("Action (1: insured, 0:non-insured)")
```
