```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('logs/outfile-2019-07-31-09-47-48.p', 'rb') as fp:
    data= pickle.load(fp)
```

```python
data
```

```python
low_lim = 28995+1000
high_lim = low_lim + 10000
```

```python
insurance_costs = np.array([x[1][:2] for x in data[low_lim: high_lim]])
```

```python
agent_actions = np.array([[0 if x in [0,1] else 1 for x in y[1][2:]] for y in data[low_lim: high_lim]])
```

```python
plt.figure(figsize=(20., 4.8))
plt.plot(range(len(insurance_costs)),insurance_costs[:,0])
plt.plot(range(len(insurance_costs)),insurance_costs[:,1])
plt.title('Insurance cost')
plt.figure(figsize=(20., 4.8))
plt.plot(range(len(insurance_costs)), agent_actions[:,0])
plt.plot(range(len(insurance_costs)), agent_actions[:,1])
plt.title("Action (1: insured, 0:non-insured)")
```
